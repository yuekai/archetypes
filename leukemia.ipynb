```python
from archetypes import Archetypes
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize as spopt
```

```python
def load_ALL_AML():

  X = np.zeros((5000,38))

  f = open('ALL_AML.txt')
  for i, ff in enumerate(f):
    X[i, :] = list(map(float, ff.split('\t')))

  return X.T
```

```python
X = load_ALL_AML()

a = Archetypes(X)
archetypes = a.findAllArchetypes()
a.weights(k=5);
```

```python
def refitH(a):
  """
  Refit H on the weights
  """

  H = np.empty((a.k,a.p))
  for j in range(a.p):
    H[:,j], _ = spopt.nnls(a.W, a.X[:,j])

  return H

def plotMetageneDensities(a,k=1):

  if k > a.k:
    raise ValueError("k must be less than or equal to the selected rank.")
    
  H = refitH(a)

  for kk in range(k):
    plt.figure()
    plt.plot(H[kk,:], alpha=0.75)
    plt.title('Density of metagene ' + str(kk+1))
    plt.show()
```

```python
plotMetageneDensities(a,k=5)
```

```python
def plotArchetypes(a,k=1):

  if k > a.k:
    raise ValueError("k must be less than or equal to the selected rank.")

  for kk in range(k):
    plt.figure()
    plt.plot(a.H[kk,:], alpha=0.75)
    plt.title('Expression profile of archetype ' + str(kk+1))
    plt.show()
```

```python
def plotMetageneExpressionProfiles(a,k=1):

  if k > a.k:
    raise ValueError("k must be less than or equal to the selected rank.")

  for kk in range(k):
    plt.figure()
    plt.plot(range(18), a.W[:18,kk], color='r', dashes='', marker='o', alpha=0.75)
    plt.plot(range(19,27), a.W[19:27,kk], color='b', marker='s', alpha=0.75)
    plt.plot(range(27,38), a.W[27:,kk], color='k', marker='^', alpha=0.75)
    plt.title('Expression profile of metagene ' + str(kk+1))
    plt.show()
```

```python
plotMetageneExpressionProfiles(a,k=5)
```
